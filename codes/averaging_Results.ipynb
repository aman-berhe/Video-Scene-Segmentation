```python
winDiff=[]
pk=[]
coverage=[]
purity=[]
recall=[]
precision=[]
for j in range(0,6):
    for i in range(j,len(resData),6):
        #print(i,j)
        #if resData['Unnamed: 0'][i]=='specteral':
        winDiff.append(resData['winDiff'][i])
        pk.append(resData['pk'][i])
        coverage.append(resData['coverage'][i])
        purity.append(resData['purity'][i])
        recall.append(resData['recall'][i])
        precision.append(resData['precision'][i])
        #else:
            #continue
    winDiff=round(sum(winDiff)/len(winDiff),3)
    pk=round(sum(pk)/len(pk),3)
    coverage=round(sum(coverage)/len(coverage),3)
    purity=round(sum(purity)/len(purity),3)
    recall=round(sum(recall)/len(recall),3)
    precision=round(sum(precision)/len(precision),3)
    F1=round((2*(recall*precision)/(recall+precision)),3)
    print('&',winDiff,'&',pk,'&',coverage, '&',purity,'&',recall,'&',precision,'&',F1)
    winDiff=[]
    pk=[]
    coverage=[]
    purity=[]
    recall=[]
    precision=[]
```
